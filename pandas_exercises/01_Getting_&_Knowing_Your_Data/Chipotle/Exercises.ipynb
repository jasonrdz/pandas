### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo['item_price'].astype

# objects
```

#### Step 13.b. Create a lambda function and change the type of item price

```python

# since it is looking at each occurence within the column, we must look at the string, so past the first index
floating  = lambda x: float(x[1:-1])

chipo.item_price = chipo.item_price.apply(floating)

chipo
```

#### Step 13.c. Check the item price type

```python
chipo['item_price'].dtype
```
